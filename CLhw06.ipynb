#### Задача 13.

Увеличить все встретившиеся в тексте четные числа на 1, а нечетные на 2, в том числе и отрицательные. Перед числом может стоять знак. Считаем, что между знаком и числом пробела нет.

```python
text13 = '4 черненьких чертёнка чертили черными чернилами -71 портрет'
c = []
b = []
for i in re.findall(r'-*\d*[24680]', text13):
    c.append(int(i) + 1)
for j in re.findall(r'-*\d*[13579]', text13):
    b.append(int(j) + 2)
print(f'четные {c},\nнечетные {b}')
```
